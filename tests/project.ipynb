```python
from office365.sharepoint.client_context import ClientContext
from office365.runtime.auth.user_credential import UserCredential

# SharePoint site and folder details
site_url = "https://jasolar.sharepoint.com/sites/Projects"
relative_folder_url = "/sites/Projects/Shared Documents/Documents and Tracking/Project Report and Report Genetation/On-going"

# Credentials (use secure method in production)
username = "deepaksureshnidagundi@jasolar.eu"
password = "Solar2023!"

# Connect to SharePoint
ctx = ClientContext(site_url).with_credentials(UserCredential(username, password))

# Get the folder
folder = ctx.web.get_folder_by_server_relative_url(relative_folder_url)
files = folder.files
ctx.load(files)
ctx.execute_query()

# List files in the folder
for file in files:
    print("File name:", file.properties["Name"])
    print("Server relative URL:", file.properties["ServerRelativeUrl"])
```
